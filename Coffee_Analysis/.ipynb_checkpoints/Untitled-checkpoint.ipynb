# Common Questions Related To Data

### 1. Top Companies producing coffees

```python
coffee["Company"].value_counts()
```

```python
def print_top_companies(df, Company):
    company_counts = df[Company].value_counts().head(5)
    print("Top 5 companies:")
    for company, count in company_counts.items():
        print(f"- {company}: {count} occurrences")

# Example usage:
# Assuming 'coffee' is the DataFrame and 'Company' is the column containing the company names
print_top_companies(coffee, 'Company')
```

### 2.Where are these companies Located?

```python
def print_top_companies_by_country(df, country_column, company_column):
    top_companies_by_country = df[country_column].value_counts().head(5)
    print("Top 5 companies by country of origin:")
    for country, count in top_companies_by_country.items():
        companies = df[df[country_column] == country][company_column].unique()
        print(f"Country: {country}")
        for company in companies:
            print(f"- {company}")
        print()


print_top_companies_by_country(coffee, 'Country of Origin', 'Company')
```
